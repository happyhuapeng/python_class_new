```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

import os

os.listdir('./')
```

```python
img = cv2.imread('imori_256x256.png')[..., ::-1]
plt.imshow(img); plt.show()
```

# dark

```python
#img_dark = (img.astype(np.float32) * 0.3).astype(np.uint8)
img_dark = (img.astype(np.float32) / 255 * 70 + 70) .astype(np.uint8)
plt.imshow(img_dark); plt.show()
Image.fromarray(img_dark).save('imori_256x256_dark.png', 'png')
```

```python
plt.hist(img_dark.ravel(), bins=255, rwidth=0.8, range=(0, 255))
plt.show()
```

# Light

```python
img_light = img_dark + 70
plt.imshow(img_light); plt.show()
Image.fromarray(img_light).save('imori_256x256_light.png', 'png')
```

```python
plt.hist(img_light.ravel(), bins=255, rwidth=0.8, range=(0, 255))
plt.show()
```

# noise

```python
np.random.seed(0)
ind = np.random.random_sample(img.shape)
ind = np.minimum(ind // 0.9, 1).astype(bool)
img_noise = img.copy()
img_noise[ind] = 0
plt.imshow(img_noise)
Image.fromarray(img_noise).save('imori_256x256_noise.png', 'png')
```

# Gamma

```python
img_gamma = np.clip(((img.astype(np.float32) / 255) ** 2.2) * 255, 0, 255).astype(np.uint8)
plt.imshow(img_gamma); plt.show()
Image.fromarray(img_gamma).save('imori_256x256_gamma.png', 'png')
```

```python
plt.hist(img_gamma.ravel(), bins=255, rwidth=0.8, range=(0, 255))
plt.show()
```
